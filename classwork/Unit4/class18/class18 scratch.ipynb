```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.plotting import autocorrelation_plot
from statsmodels.tsa.stattools import acf, acovf

plt.style.use('fivethirtyeight')
%matplotlib inline

df = pd.read_csv('data/births.csv', index_col='Date', parse_dates = True)
df.head()
```

```python
df.plot(title = 'Daily Births California, 1959');
```

```python
X_t = df[:-1]['Births'] - df['Births'].mean()
X_t.head()
```

```python
X_t1 = df[1:]['Births'] - df['Births'].mean()
X_t1.head()
```

```python
X_cov_t1 = np.sum(X_t.values * X_t1.values) * (1 / len(df))
X_cov_t1 / df['Births'].var()
```

```python
X_t = df[:-10]['Births'] - df['Births'].mean()
X_t10 = df[10:]['Births'] - df['Births'].mean()
X_cov_t10 = np.sum(X_t.values * X_t10.values) * (1 / len(df))
X_cov_t10 / df['Births'].var()
```

```python
acf(df)
```

```python
acovf(df)
```
