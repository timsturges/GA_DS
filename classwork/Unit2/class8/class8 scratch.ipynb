```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/titanic.csv')
```

```python
df.head(10)
```

```python
df['Sex'].value_counts().plot(kind = 'bar');
```

```python
df['Embarked'].head(10)
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind = 'barh');
```

```python
df.groupby('Sex')[['Age','Fare']].mean().plot(kind = 'bar',
                                              title = 'Average Age and Fare by Passenger Gender');
```

```python
df['Embarked'].value_counts().plot(kind = 'pie');
```

```python
df['Age'].plot(kind = 'hist', 
               bins = 20);
```

```python
df.hist(column = 'Age', 
        by = 'Embarked', 
        layout = (1,3), sharey = True);
```

```python
df['Age'].plot(kind = 'kde');
```

```python
df.plot('Age','Fare', kind = 'scatter');
```

```python
df.plot('Age','Fare', kind = 'hexbin');
```

```python
fig, ax = plt.subplots(1, 2, 
                       sharex = True, 
                       sharey = True, 
                       figsize = (12,4))
df[df['Sex'] == 'female'].plot('Age',
                               'Fare',
                               kind = 'scatter', 
                               ax = ax[0], 
                               color = 'red', 
                               label = 'Women')
df[df['Sex'] == 'male'].plot('Age',
                             'Fare',
                             kind = 'scatter', 
                             ax = ax[1], 
                             color = 'blue', 
                             label = 'Men');
```

```python
import seaborn as sns
```

```python
sns.catplot(x = 'Pclass',
            y = 'Age',
            data = df);
```

```python
sns.catplot(x = 'Pclass',
            y = 'Age',
            data = df, 
            kind = 'box');
```

```python
sns.catplot(x = 'Pclass',
            y = 'Age',
            data = df, 
            kind = 'swarm',
            hue = 'Sex');
```

```python
sns.catplot(x = 'Pclass',
            y = 'Fare',
            data = df, 
            kind = 'swarm',
            col = 'Sex');
```

```python
sns.catplot(x = 'Pclass',
            y = 'Fare',
            data = df, 
            kind = 'swarm',
            row = 'Sex');
```

```python
sns.catplot(x = 'Pclass',
            y = 'Age',
            data = df, 
            kind = 'swarm',
            row = 'Survived',
            col = 'Sex');
```

```python
sns.catplot(x = 'Pclass',
            y = 'Survived',
            data = df, 
            kind = 'bar',
            hue = 'Sex')
plt.title('Survival Rate by Passenger Class and Sex');
```

```python
sns.lmplot(x = 'Pclass', 
           y = 'Survived', 
           data = df);
```

```python
sns.distplot(df['Age'].fillna(df['Age'].median()));
```

```python
sns.jointplot(x = 'Age', y = 'Fare', data = df[df['Sex'] == 'male']);
```

```python
ax = sns.distplot(df[df['Sex'] == 'female']['Age'].dropna(), label = 'Women')
sns.distplot(df[df['Sex'] == 'male']['Age'].dropna(), label = 'Men', ax = ax)
plt.legend(loc = 'best');
```

```python
sns.pairplot(df);
```
