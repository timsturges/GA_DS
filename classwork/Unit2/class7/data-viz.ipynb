### Data Visualization With Pandas, Matplotlib, & Seaborn

The major chart types used in graphing data with Python.

 - **Matplotlib**: The major library that others are built off of
 - **Pandas**: Has built in matplotlib functions that allow you to graph based off of a dataframe
 - **Seaborn**: User-friendly version of matplotlib, has a number of custom chart types that are very handy.
 
A really important detail is getting the correct dataframe for each graph!

```python
# load in the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('../data/titanic.csv')
```

```python
df.head()
```

```python
# we can take this dataframe view
df.Sex.value_counts()
```

```python
# and turn it into this:
df.Sex.value_counts().plot(kind='bar')
```

### Categorical Data

 - has the generic form:  
  - category value -- numeric value
  - charts that can be created:
   - bar
   - barh
   - boxplot
   - pie

```python
# need to have information situation correctly
df['Embarked'].head(10)
```

```python
# the following will not chart well
df['Embarked'].plot(kind='bar')
```

```python
# let's take a look at another example
df.groupby('Sex')['Survived'].mean()
```

```python
# could chart it in the following way
df.groupby('Sex')['Survived'].mean().plot(kind='barh')
```

```python
from IPython.display import Image
Image(filename='plot.jpg')
```

```python
# correct code would look like this
df.groupby('Sex')[['Age', 'Fare']].mean().plot(kind='bar')
```

```python
# this is how you make a histogram
df['Age'].plot(kind='box')
```

```python
df.boxplot(column='Age', by='Embarked', grid=False)
```

```python
# pie chart
df['Embarked'].value_counts().plot(kind='pie')
```

```python
# pie chart
df['Embarked'].value_counts().plot(kind='pie')
```

### Quantitative Data

 - Either charting a single numeric variable, or numeric variables against one another
  - Major chart types:
   - histogram
   - scatter plot
   - kde
   - hexbin

```python
# histogram
df['Age'].plot(kind='Hist')
```

```python
# a few helpful options
df['Age'].plot(kind='Hist', bins=20)
```

```python
# can also plot a histogram across an entire dataframe, and specify options
df.hist(column='Age', by='Embarked')
```

```python
# can specify the layout of columns and rows
df.hist(column='Age', by='Embarked', layout=(1, 3))
```

```python
# and the figure size
df.hist(column='Age', by='Embarked', layout=(1, 3), figsize=(12, 4))
```

```python
# you can also force subplots to share an x and y axis
df.hist(column='Age', by='Embarked', layout=(1, 3), figsize=(12, 4), sharey=True)
```

## A Few More Helpful Charts......

```python
# like a histogram, only its graphed based on density estimates
df['Age'].plot(kind='kde')
```

```python
# create a scatter plot
df.plot('Age', 'Fare', kind='Scatter', title='Age vs. Fare')
```
